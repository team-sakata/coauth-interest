## coauthor_works

```python
def coauthor_works_null_year(id):
    works = authors_valid['eid'][authors_valid['authid']==id]
    selected_papers = year_df[year_df.index.isin(works.values)]
    selected_papers["year"] = selected_papers["year"].sample(frac=1).values
    selected_papers = selected_papers.sort_values(by="year")
    selected_papers = selected_papers.assign(seq=list(range(0,len(selected_papers))))
    coauthors = authors_valid[authors_valid['eid'].isin(works)].query(f"authid!={id}")
    
    df_seqyear = pd.merge(coauthors,selected_papers, how='left', on='eid').sort_values(by='seq')
    df_seqyear_multi = df_seqyear.groupby(by='authid', group_keys=True).filter(lambda x: len(x) > 1)

    if len(df_seqyear_multi)<1: return None
    df_seqyear_delta = df_seqyear_multi.groupby(by='authid', group_keys=True)[['year','seq']] \
    .apply(lambda x:  x - x.shift(1)) \
    .dropna().astype(int).reset_index()
    
    df_seqyear_delta = pd.merge(df_seqyear_delta, df_seqyear["eid"], how="left", left_on="level_1", right_index=True).drop(columns=["level_1"])
    df_seqyear_delta["source_authid"] = id
    df_seqyear_delta["previous_work"] = df_seqyear_delta.groupby("authid").shift(1)["eid"]

    df_seqyear_delta = df_seqyear_delta.dropna()
    return df_seqyear_delta


coauthor_works_null_year(6506905929)

big_author_trimmed = pd.read_pickle("./result/big_author_trimmed_PM.pickle")
sample_author=random.sample(list(big_author_trimmed), len(list(big_author_trimmed)))
```

```python
# random sampling
iter = 1000
for j in range(3,10):
    res = pd.DataFrame()
    for i in range(iter):
        res = pd.concat([res, coauthor_works_null_year(sample_author[i+j*iter])],axis=0)
    res.to_pickle(f"./result/PM_null_year{j:02}.pickle")
res
```

```python
import numpy as np
import math
from scipy.optimize import curve_fit

# ff =res.query("is_ff==True")
data = res_shuffled[res_shuffled["year"]>0]["year"].values
# data = res["year"].values
hist, bins = plotter(data,title="coauthorship interval",c='blue',s=25, u= "year", l="Physics & Materials Science",count="number", plot="power-law")

x_data = np.log(bins[:15])
y_data = np.log(hist[:15])

# 直線の方程式を定義します
def linear_func(x, m, b):
    return m * x + b

# 最小二乗法を使用してパラメータをフィットします
params, covariance = curve_fit(linear_func, x_data, y_data)

# フィットされたパラメータを表示します
m_fit, b_fit = params

# 確率密度関数を定義します
def power_law(x, alpha, xmin):
    return math.exp(xmin)*x**alpha
    # return (alpha - 1) / xmin * (x / xmin) ** (-alpha)
    
x = np.linspace(min(data), 200, 50)
plt.plot(x, power_law(x, m_fit,b_fit),'--',c='black', linewidth=1, label='Power Law Fit')

print("傾き (m):", m_fit)
print("切片 (b):", b_fit)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import poisson

data = res.query("year>0")['year'].values
bins = [i for i in range(int(max(data))+1)]
bins_plot = np.add(bins,(bins[1] - bins[0])/2)[:-1]
hist,bins = np.histogram(data, bins=bins)
hist = np.cumsum(hist)/len(data)

plt.scatter(bins[:-1], hist,c='blue', label="empirical")

def fit_function(k, lamb):
    '''poisson function, parameter lamb is the fit parameter'''
    return poisson.cdf(k, lamb)


# fit with curve_fit
parameters, cov_matrix = curve_fit(fit_function, bins_plot, hist)

# plot poisson-deviation with fitted parameter
x_plot= bins

plt.plot(
    x_plot,
    fit_function(x_plot, *parameters),
    linestyle='-',c='orange',
    label='Poisson-CDF Fit',
)
plt.ylim(0.001,1)
# plt.yscale('log')
# plt.xscale('log')
plt.legend()
plt.show()
parameters
```

```python
from scipy.stats import expon
from scipy.stats import pareto

data = res.query("year>0 and year<15")['year'].values
b, loc2, scale2 = pareto.fit(data, 2.4, loc=0, scale=1)

bins = [i for i in range(int(max(data))+1)]
bins_plot = np.add(bins,1)
hist,_ = np.histogram(data, bins=bins)
hist = 1- np.cumsum(hist)/len(data)

plt.scatter(bins_plot[:-1], hist,edgecolor='blue', c='white',  label="empirical")

plt.plot(
    bins,
    pareto.sf(bins, b=2.5),
    linestyle='-',c='red',
    label='Power law Fit',
)

plt.ylim(hist[-2],2)
plt.yscale('log')
plt.xscale('log')
plt.legend()

b, loc2, scale2
```
