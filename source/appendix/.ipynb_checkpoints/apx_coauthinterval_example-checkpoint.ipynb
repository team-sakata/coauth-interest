# Barabasi example

```python
import pandas as pd
df = pd.read_csv("coauthor_timedistribution.csv")
df = df.iloc[1:,1:-1]
df = df.astype('string')
df
```

```python
import pandas as pd

def get_natural_number_intervals(series):
    intervals = []
    count_zeros = 1

    for value in series:
        if value == "0":
            count_zeros += 1
        elif count_zeros!="0":
            intervals.append(count_zeros)
            count_zeros = 1

    return intervals

series = pd.Series(["0","1","1","0","0","1"])
print(series.values)
result = get_natural_number_intervals(series)
print(result)
```

```python
import pandas as pd

def find_first_authorship(series):
    for i, value in enumerate(series):
        if value == "1":
            return True
        elif value != "0":
            return False
    return False

# # Example 1
# series1 = ["0", "1", "0", "2", "3", "0", "LAST", "1"]
# result1 = find_first_authorship(series1)
# print(result1)  # Output: 2

# # Example 2
# series2 = ["0", "LAST", "1", "2", "3", "0", "LAST", "0"]
# result2 = find_first_authorship(series2)
# print(result2)  # Output: 3

test =  pd.DataFrame([["0", "1", "0", "2", "3", "0", "LAST", "1"],["0", "LAST", "1", "2", "3", "0", "LAST", "0"]])
test.apply(lambda x :find_first_authorship(x), axis=1)
```

```python
first = df.apply(lambda x: find_first_authorship(x), axis=1)
first
```

```python
df['res'] = [None] * len(df)
for (i,d) in df.iterrows():
    result = get_natural_number_intervals(d)
    result.pop()
    df.at[i,"res"] = result
df
```

```python
df_first = df[first]
df_notfirst = df[~first]
```

```python
from util import plotter
import matplotlib.pyplot as plt
import numpy as np

res = df["res"].explode()
res_first = df_first["res"].explode()
res_notfirst = df_notfirst["res"].explode()

bin = res.max()

# ヒストグラムの描画
# counts, bins, _ = plt.hist(res, bins=bin, color='blue', alpha=0.4, density=True, label='データ分布')
counts_first, bins_first, _ = plt.hist(res_first, bins=bin, color='red', alpha=0.2, density=True, label='First-first')
counts_notfirst, bins_notfirst, _ = plt.hist(res_notfirst, bins=bin, color='blue', alpha=0.2, density=True, label='non-first')

# 累積分布関数 (CDF) の計算
# cumulative = np.cumsum(counts * np.diff(bins))
cumulative_first = np.cumsum(counts_first * np.diff(bins_first))
cumulative_notfirst = np.cumsum(counts_notfirst * np.diff(bins_notfirst))


# 相補累積分布関数 (CCDF) の計算
# ccdf = 1 - cumulative
ccdf_first = 1 - cumulative_first
ccdf_notfirst = 1 - cumulative_notfirst

# CCDFの描画
# plt.plot(bins[:-1], ccdf, color='black', label='相補累積分布関数 (CCDF)')
plt.plot(bins_first[:-1], ccdf_first, color='red', label='First-first-CCDF')
plt.plot(bins_notfirst[:-1], ccdf_notfirst, color='blue', label='nonfirst-CCDF')
plt.xscale('log')
plt.yscale('log')
plt.ylim(0.001,1)
# plt.legend()

plt.title('Barabasi Coauthorship intervals')
plt.xlabel('Coauthorship interval [papers]')
plt.ylabel('Probability')

plt.show()
```
