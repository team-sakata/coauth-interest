```python
interval = long.groupby("eid")["year"].mean()
long_term_impact = pd.merge(interval,numcited_df, how="left",left_on="eid",right_index=True)
long_term_valid = long_term_impact.dropna().sort_values("year")
long_term_valid = long_term_valid[long_term_valid["source"] < long_term_valid["source"].quantile(0.99)]
long_term_valid
```

```python
x = np.arange(len(long_term_valid))/len(long_term_valid)
y = np.cumsum(long_term_valid["source"]) / np.sum(long_term_valid["source"])

plt.plot(x,y)
plt.plot([0, 1], [0, 1], '--',color='gray', label='Line of equality')
plt.xlabel("year")
```

```python
bins = np.arange(1,30)
impact = []
for i in range(len(bins)-1):
    impact.append(long_term_valid.query(f"year>={bins[i]} and year<{bins[i+1]}")["source"].mean())
bins_plot = (bins[:-1]+bins[1:])/2

plt.plot(bins_plot, impact)
```
