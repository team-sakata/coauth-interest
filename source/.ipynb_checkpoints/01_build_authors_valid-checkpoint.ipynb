```python
from dotenv import load_dotenv
load_dotenv()

import os
data_path = os.getenv('DATA_PATH')
result_path = os.getenv('RESULT_PATH')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import random
from scipy.stats import expon
from scipy.stats import pareto
import os

from util import plotter
```

## Valid Authors 

```python
year_df = pd.read_pickle(data_path + "/paper_detail/year.pickle")
year_df = pd.DataFrame(year_df.sort_values())
print(year_df.info())
year_df.head()
```

```python
author_df = pd.read_pickle(data_path + "/id_names/author_df.pickle")
print(author_df.info())
author_df.head()
```

```python
# drop "et al.", "No NAME" or suspicious authids from author_df, whoose id is of 0, 1, 2
authors_valid = author_df[~author_df['authid'].isin([0,1,2])] \
    .drop_duplicates(subset=["eid","authid","authseq"]) \
    .astype({'eid':'int','authid': 'int','authseq':'int'})
print(authors_valid.shape[0])
# the change in the number of entries 327,280,829 → 289,839,055 (8min)

# drop not article (article, book, book chapter, conference paper, data paper, review)
doctype_df = pd.read_pickle(data_path + "/paper_detail/doctype.pickle")
authors_valid = pd.merge(authors_valid, doctype_df, how="left", on="eid")
temp_isin = authors_valid["doctype"].isin(["ar", "bk", "ch", "cp","dp", "re"])
authors_valid = authors_valid[temp_isin]
print(authors_valid.shape[0])
# → 277,724,637

# drop big project
team_size = authors_valid[["authid","eid"]].groupby(by="eid").count()
large_paper = team_size.query("authid>30").index.values

authors_valid = authors_valid[~authors_valid["eid"].isin(large_paper)]
print(authors_valid.shape[0])
# → 263,309,553
```

```python
authors_valid.to_pickle(result_path +"/authors_valid.pickle")
```

# Partitioned Valid Authors 

```python
partition_df = pd.read_pickle(data_path + "/paper_detail/partition.pickle")
partition_df.info() # 66,685,170

disciplines = {"So":0,"CS":1,"EB":2,"PM":3,"Ne":5,"Im":6,"Nt":7,"Ch":9,"Ps":11,"Mol-Md":4,"Cli-Md":13,"AB":16,"As":17,"Ma":23}
partitions = pd.DataFrame({"name":disciplines.keys()},index=disciplines.values())
partitions.info()
```

```python
authors_valid = pd.read_pickle(result_path + "/authors_valid.pickle") 

authors_valid_partitioned = pd.merge(authors_valid,partition_df, how="left",left_on="eid", right_index=True)\
    .dropna(subset=["partition"])
authors_valid_partitioned.info() # 244,736,690
```

```python
def build_authors_valid_partitioned(partition, name):
    temp = authors_valid_partitioned[authors_valid_partitioned["partition"]==partition]
    temp.to_pickle(result_path + f"/authors_valid_{name}.pickle")
    return temp.shape[0]

for d in partitions["name"].values:
    authors_count = build_authors_valid_partitioned(d[0],d[1]["name"])
    partitions.loc[d,"authors_count"] = authors_count
    
# build_authors_valid_partitioned(0,"So")     # 0: Social Science         17,088,254 rows × 8 columns
# build_authors_valid_partitioned(1,"CS")     # 1: Computer Science       23,210,192 rows × 8 columns
# build_authors_valid_partitioned(2,"EB")     # 2: Environmental Biology  20,679,304 rows × 8 columns
# build_authors_valid_partitioned(3,"PM")     # 3: Physics and Materials Science  22,084,638 rows × 8 columns
# build_authors_valid_partitioned(5,"Ne")     # 5: Neurology              13,429,168 rows × 8 columns
# build_authors_valid_partitioned(6,"Im")     # 6: Immunology             20,679,304 rows × 8 columns
# build_authors_valid_partitioned(7,"Nt")     # 7: Nanotechnology         13,429,168 rows × 8 columns
# build_authors_valid_partitioned(9,"Ch")     # 9: Chemistry              14,575,853 rows × 8 columns
# build_authors_valid_partitioned(11,"Ps")    #11: Psychology             8,975,764 rows × 8 columns
# build_authors_valid_partitioned(4,"Mol-Md") # 4: Molecular Medicine     25,523,029 rows × 8 columns
# build_authors_valid_partitioned(13,"Cli-Md")#13: Clinical Medicine      9,880,930rows × 8 columns
# build_authors_valid_partitioned(16,"AB")    #16: Antibiotic Research    8,067,684rows × 8 columns
# build_authors_valid_partitioned(17,"As")    #17: Astrophysics           5,866,880rows × 8 columns
# build_authors_valid_partitioned(23,"Ma")    #23: Mathematics            21,785rows × 8 columns
```

## Winsorized Big Author

```python
import math
import pickle
from scipy.optimize import curve_fit

percentile = 0.01
for d in partitions.iterrows():
    authors_valid = pd.read_pickle(result_path + f"/authors_valid_{d[1]["name"]}.pickle")
    print("confirm unique partitions in the file:", authors_valid["partition"].unique())
    team_size = authors_valid[["authid","eid"]].groupby(by="eid").count().max()
    print("confirm max team size: ",max(team_size["authid"].values),"authors in a single paper")
    
    coauthorship_count = authors_valid[['eid', 'authid']].groupby(by='authid').count()
    big_author = coauthorship_count.rename({"eid":"n_papers"},axis=1)

    data = big_author["n_papers"].values
    hist, bins = plotter(data,u="papers", count="probability",plot="power-law",title="Distribution # Authorship",l=d[1]["name"],xlabel="# papers produced", ylabel="probability authors")
    
    x_data = np.log(bins[:50])
    y_data = np.log(hist[:50])
    
    def linear_func(x, m, b):
        return m * x + b
    params, covariance = curve_fit(linear_func, x_data, y_data)
    m_fit, b_fit = params

    def power_law(x, alpha, xmin):
        return math.exp(xmin)*x**alpha
    
    x = np.linspace(min(data), 100, 100)
    
    plt.clf()
    plt.plot(x, power_law(x, m_fit,b_fit),'--',c='black', linewidth=1, label=f'Power Law Fit x^{round(m_fit,2)} * {round(b_fit,2)}')
    plt.legend()
    plt.savefig(result_path+f"/authorship_{d[1]["name"]}.png")
    
    print("傾き (m):", m_fit)
    print("切片 (b):", b_fit)

    plt.show()
    
    # drop top 1% and bottom 1% of the distribution
    print(f"trim top and bottom {int(percentile*100)}% in the data")
    threshold_top = big_author["n_papers"].quantile(1-percentile)
    threshold_bottom = big_author["n_papers"].quantile(percentile)
    print(big_author.query(f"n_papers<{threshold_bottom}").count().values, "authors have less than ",threshold_bottom," papers")
    print(big_author.query(f"n_papers>{threshold_top}").count().values, "authors have more than ",threshold_top," papers")
    
    big_author_trimmed = big_author.query(f"{threshold_bottom}<n_papers<{threshold_top}").index.values
    
    random.seed(a=13)
    sample_author=random.sample(list(big_author_trimmed), len(list(big_author_trimmed)))
    
    file_path = result_path + f"/sample_authors_{d[1]["name"]}.pickle"
    
    with open(file_path, 'wb') as file:
        pickle.dump(sample_author, file)
    print("there are",big_author_trimmed.shape[0],"valid authors within the thresholds")
    print(big_author.sort_values(by="n_papers",ascending=False).head(10))
    print("---------")
    
```
